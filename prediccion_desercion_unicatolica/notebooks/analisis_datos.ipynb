```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pathlib as p
```

```python
path = p.Path('../data/processed/data_set.csv')
path.exists()
```

```python
df = pd.read_csv(path)
df.head()
```

```python
plt.style.use('seaborn-v0_8')

sns.histplot(df,x='edad',bins=10)
plt.ylabel('cantidad estudiantes')
```

```python
sns.boxplot(df,x='edad')
```

```python
sns.violinplot(data=df['edad'])
```

```python
genero_tipo = ['Masculino','Femelino']

genero_cantidad = df['genero'].value_counts()
print(genero_cantidad)

plt.pie(genero_cantidad,labels=genero_tipo,autopct='%1.1f%%')
plt.legend()
plt.show()
```

```python
sns.countplot(df,x='estrato',order=[1,2,3,4,5])
```

```python
sns.countplot(data=df,x='residencia',order=[0,1,2])
plt.ylabel('cantidad de estudiantes')
```

```python
periodos = df['periodo_inicio'].unique().sort()

sns.countplot(data=df,x='periodo_inicio',order=periodos)
plt.xticks(rotation=60)
plt.ylabel('cantidad estudiantes')
plt.xlabel('estudiantes inscritos por periodo')
```

```python
desertor_tipo = ['no desertor','desertor']
desertor_cantidad = df['desertor'].value_counts()

plt.pie(desertor_cantidad,labels=desertor_tipo,autopct='%1.1f%%')
plt.title('DeserciÃ³n entre 2018 - 2024')
plt.legend()
plt.show()
```
